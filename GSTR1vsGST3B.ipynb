```python
from google.colab import files
uploaded_r1 = files.upload()
```

```python
monthly_totals = []

for filename in uploaded_r1.keys():
    print(f"\n📂 Processing {filename}")

    xls_r1 = pd.ExcelFile(filename)

    if "b2b, sez, de_inv" not in xls_r1.sheet_names:
        print(f"⚠️ Skipping {filename} - No 'b2b, sez, de_inv' sheet")
        continue

    # Read sheet
    df_r1 = pd.read_excel(xls_r1, sheet_name="b2b, sez, de_inv", header = 3)

    # 🔎 Step 1: Show original columns (before cleaning)
    print("👉 Original columns:", df_r1.columns.tolist())

    # 🔧 Step 2: Clean column names
    df_r1.columns = df_r1.columns.str.strip().str.lower()

    # 🔎 Step 3: Show cleaned columns
    print("👉 Cleaned columns:", df_r1.columns.tolist())

    # Step 4: Pick needed ones
    needed = ["invoice date", "taxable value", "integrated tax", "central tax", "state/ut tax"]
    df_r1 = df_r1[[c for c in needed if c in df_r1.columns]]

    # Step 5: Process
    df_r1["invoice date"] = pd.to_datetime(df_r1["invoice date"], errors="coerce", dayfirst = True)
    df_r1["Month"] = df_r1["invoice date"].dt.to_period("M")
    monthly_sum_r1 = df_r1.groupby("Month")[["taxable value", "integrated tax", "central tax", "state/ut tax"]].sum().reset_index()
    monthly_totals.append(monthly_sum_r1)
```

```python
final_r1 = pd.concat(monthly_totals, ignore_index=True)
final_r1 = final_r1.groupby("Month")[["taxable value", "integrated tax", "central tax", "state/ut tax"]].sum().reset_index()
# Step 5: Sort by Month
final_r1 = final_r1.sort_values(by="Month")
```

```python
totals_r1 = final_r1.drop(columns="Month").sum()
totals_r1_row = pd.DataFrame([{"Month": "TOTAL (FY 2020-21)", **totals_r1.to_dict()}])
final_r1 = pd.concat([final_r1, totals_r1_row], ignore_index=True)
```

```python
from pandas.tseries.offsets import MonthEnd
```

```python
month_lookup = {
    "Jan": 1, "Feb": 2, "Mar": 3,
    "Apr": 4, "May": 5, "Jun": 6,
    "Jul": 7, "Aug": 8, "Sep": 9,
    "Oct": 10, "Nov": 11, "Dec": 12
}
```

```python
monthly_totals_3b = []
for filename in uploaded_3b.keys():
    print(f"\n📂 Processing {filename}")
    xls_3b = pd.ExcelFile(filename)

    reqd_row = pd.read_excel(xls_3b, sheet_name="GSTR-3B", skiprows = 6, nrows = 1, header=None)
    month_abbr = str(reqd_row.iloc[0, 4]).strip()[:3]
    m = month_lookup.get(month_abbr)
    year = 2021 if m in [1, 2, 3] else 2020
    tax_period = pd.Period(year=year, month=m, freq="M")

    print("Tax period detected:", tax_period)
    df_3b = pd.read_excel(xls_3b, sheet_name="GSTR-3B", header=15, nrows=5)
    df_3b.columns = df_3b.columns.str.strip().str.lower()
    df_3b = df_3b.rename(columns={"total taxable value": "taxable value"})
    print("👉 Cleaned columns:", df_3b.columns.tolist())

    needed_3b = ["taxable value", "integrated tax", "central tax", "state/ut tax"]
    df_3b = df_3b[[c for c in needed_3b if c in df_3b.columns]]

    df_3b["Month"] = tax_period

    monthly_sum_3b = df_3b.groupby("Month")[needed_3b].sum().reset_index()
    monthly_totals_3b.append(monthly_sum_3b)

final_3b = pd.concat(monthly_totals_3b, ignore_index=True)
final_3b = final_3b.groupby("Month")[needed_3b].sum().reset_index()
final_3b = final_3b.sort_values(by="Month")

totals_3b = final_3b.drop(columns="Month").sum()
totals_3b_row = pd.DataFrame([{"Month": "TOTAL (FY 2020-21)", **totals_3b.to_dict()}])
final_3b = pd.concat([final_3b, totals_3b_row], ignore_index=True)


r1_clean = final_r1[final_r1["Month"] != "TOTAL (FY 2020-21)"]
r3b_clean = final_3b[final_3b["Month"] != "TOTAL (FY 2020-21)"]

merged = pd.merge(r1_clean, r3b_clean, on="Month", how="outer", suffixes=("_R1", "_R3B"))
comparison_df = pd.DataFrame()
comparison_df["Month"] = merged["Month"]
```

```python
for col in ["taxable value", "integrated tax", "central tax", "state/ut tax"]:
    r1_col = f"{col}_R1"
    r3b_col = f"{col}_R3B"
    diff_col = f"{col}_Diff(R3B-R1)"
    comparison_df[diff_col] = merged[r3b_col] - merged[r1_col]
# Add TOTAL row for comparison
totals_comp = comparison_df.drop(columns="Month").sum()
totals_comp_row = pd.DataFrame([{"Month": "TOTAL (FY 2020-21)", **totals_comp.to_dict()}])
comparison_df = pd.concat([comparison_df, totals_comp_row], ignore_index=True)
```

```python
output_filename = "GST_R1_vs_R3B_Comparison.xlsx"

with pd.ExcelWriter(output_filename, engine="xlsxwriter") as writer:
    final_r1.to_excel(writer, index=False, sheet_name="GST-R1")
    final_3b.to_excel(writer, index=False, sheet_name="GST-R3B")
    comparison_df.to_excel(writer, index=False, sheet_name="Comparison")

    # Formatting
    workbook  = writer.book
    for sheet in ["GST-R1", "GST-R3B", "Comparison"]:
        worksheet = writer.sheets[sheet]
        bold_fmt = workbook.add_format({"bold": True})
        last_row = len(comparison_df) if sheet=="Comparison" else (
            len(final_r1) if sheet=="GST-R1" else len(final_3b)
        )
        worksheet.set_row(last_row, None, bold_fmt)

files.download(output_filename)
```
