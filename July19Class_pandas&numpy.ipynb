```python
frame1 = pd.DataFrame({'ID':["ball", 'pencil', 'pen', 'eraser', 'mug'],
                       'Price':['20.20', '2.50', '10.50', '5.00', '30.20'],
                       'Colour':['blue', 'brown', 'blue', 'white', 'orange']})
frame1
```

```python
frame2 = pd.DataFrame({'ID':['pencil', 'pen', 'ball', 'mug'],
                       'brand':['UFH', 'FMF', 'FHCJ', 'FHD']})
frame2
```

```python
pd.merge(frame1, frame2)
```

```python
pd.merge(frame1, frame2, on = 'ID')
```

```python
frame2
```

```python
frame2.columns = ['brand', 'sid']
frame2
```

```python
pd.merge(frame1, frame2, left_on = 'ID', right_on = 'sid')
```

```python
frame2.columns = ['brand', 'ID']
frame2
```

```python
pd.merge(frame1, frame2, on = 'ID', how = 'outer')
```

```python
pd.merge(frame1, frame2, on = ['brand', 'ID'], how = 'outer')
```

```python
frame2.columns = ['ID', 'brand']
frame2
```
